```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("student_data.csv")
df
```

```python
from sklearn.model_selection import train_test_split
x=df[["StudyHours"]]
y=df["ExamScore"]
plt.scatter(x,y,color="green")
plt.xlabel("StudyHours")
plt.ylabel("ExamScore")
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
plt.scatter(x_train["StudyHours"],y_train,color="blue",label="Train Data")
plt.scatter(x_test["StudyHours"],y_test,color="red",label="Test Data")
plt.plot(x_train,model.predict(x_train),color="black",label="Regression")
plt.xlabel("StudyHours")
plt.ylabel("ExamScore")
plt.legend()
plt.show()
```

```python
print(f"Mean Squared Error : {mse}")
print(f" R-Error : {r2}")
```
