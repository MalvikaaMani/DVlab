```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.read_csv("sales (part A).csv")
df
```

```python
model=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x1=df[["AdvertisingExpenditure"]]
y=df["SalesRevenue"]
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.3,random_state=42)
model.fit(x_train,y_train)
print("Analysis for Advertising Expenditure")
print(f"Coefficient : {model.coef_}")
print(f"Intercept : {model.intercept_}")
plt.scatter(x_train,y_train,color="blue",label="Training data")
plt.scatter(x_test,y_test,color="red",label="Testing Data")
plt.plot(x_train,model.predict(x_train),color="black",label="Regression")
plt.title("AdvertisingExpenditure vs SalesRevenue")
plt.xlabel("AdvertisingExpenditure")
plt.ylabel("SalesRevenue")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x2=df[["StoreLocation"]]
x_train,x_test,y_train,y_test=train_test_split(x2,y,test_size=0.3,random_state=42)
model.fit(x_train,y_train)
print("Analysis for Store Location")
print(f"Coefficient : {model.coef_}")
print(f"Intercept : {model.intercept_}")
plt.scatter(x_train,y_train,color="blue",label="Training data")
plt.scatter(x_test,y_test,color="red",label="Testing Data")
plt.plot(x_train,model.predict(x_train),color="black",label="Regression")
plt.title("StoreLocation vs SalesRevenue")
plt.xlabel("StoreLocation")
plt.ylabel("SalesRevenue")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x3=df[["Competition"]]
y=df["SalesRevenue"]
x_train,x_test,y_train,y_test=train_test_split(x3,y,test_size=0.3,random_state=42)
model.fit(x_train,y_train)
print("Analysis for Competition")
print(f"Coefficient : {model.coef_}")
print(f"Intercept : {model.intercept_}")
plt.scatter(x_train,y_train,color="blue",label="Training data")
plt.scatter(x_test,y_test,color="red",label="Testing Data")
plt.plot(x_train,model.predict(x_train),color="black",label="Regression")
plt.title("Competition vs SalesRevenue")
plt.xlabel("Competition")
plt.ylabel("SalesRevenue")
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
x=df[["AdvertisingExpenditure","StoreLocation","Competition"]]
y=df["SalesRevenue"]
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
f_stat=model.fvalue
t_stat_advertising=model.tvalues["AdvertisingExpenditure"]
t_stat_store=model.tvalues["StoreLocation"]
t_stat_competition=model.tvalues["Competition"]

print(f"F-statistic for the model : {f_stat}")
print(f"t-statistic for Advertsing Expenditure : {t_stat_advertising}")
print(f"t-statistic for Store Location : {t_stat_store}")
print(f"t-statistic for competition : {t_stat_competition}")
```
